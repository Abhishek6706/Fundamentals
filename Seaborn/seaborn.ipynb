### Why Seaborn if Matplotlib Already Exists?

Seaborn is built **on top of Matplotlib** to make visualization **easier, more attractive, and more statistical**.

#### Core Difference

| Feature              | Matplotlib                     | Seaborn                          |
|---------------------|--------------------------------|----------------------------------|
| Level               | Low-level                      | High-level                       |
| Code length         | More code                      | Less code                        |
| Default styling     | Basic                          | Beautiful                        |
| Statistical plots   | Limited                        | Strong support                   |
| DataFrame support   | Manual handling                | Direct support                   |

#### Why Seaborn was created

Matplotlib:
- Gives **full control**  
- But requires **more code and setup**

Seaborn:
- Simplifies plotting  
- Adds **better defaults + statistical features**


#### Example Difference (Conceptual)

- Matplotlib → you manually:
  - set colors  
  - labels  
  - grouping  

- Seaborn → automatically:
  - groups data  
  - applies themes  
  - handles categories  


#### Key Advantages of Seaborn

- Built-in **themes and styling**
- Works directly with **Pandas DataFrames**
- Supports:
  - categorical plots  
  - distributions  
  - regression plots  
- Better for **EDA (Exploratory Data Analysis)**  


#### When to Use What?

| Use Case                    | Preferred Library |
|----------------------------|------------------|
| Quick analysis (EDA)       | Seaborn          |
| Complex customization      | Matplotlib       |
| Statistical visualization  | Seaborn          |
| Fine-grained control       | Matplotlib       |


#### Important Insight

Seaborn = Matplotlib + Better Defaults + Statistics

### Figure-level vs Axes-level Functions (Seaborn)


#### Axes-level Functions

- Work on a **single plot (axes)**
- Plots are drawn on an **existing matplotlib axis (`ax`)**

Examples:
- `sns.scatterplot()`
- `sns.lineplot()`
- `sns.histplot()`
- `sns.boxplot()`


#### Characteristics

| Feature        | Axes-level |
|---------------|------------|
| Works on      | Single plot |
| Needs `ax`    | Optional (can pass manually) |
| Flexibility   | High (can combine multiple plots) |
| Output        | Returns `Axes` object |


#### Figure-level Functions

- Create **entire figure automatically**
- Can create **multiple subplots (facets)**

Examples:
- `sns.relplot()`
- `sns.catplot()`
- `sns.displot()`
- `sns.lmplot()`


#### Characteristics

| Feature        | Figure-level |
|---------------|--------------|
| Works on      | Whole figure |
| Creates       | Multiple plots |
| Flexibility   | Less control |
| Output        | Returns `FacetGrid` |


#### Key Difference

| Aspect            | Axes-level            | Figure-level         |
|-------------------|----------------------|----------------------|
| Scope             | One plot             | Entire figure        |
| Subplots          | Manual               | Automatic            |
| Control           | More control         | Less control         |
| Use case          | Custom plotting      | Quick multi-plots    |


#### Key Insight

```text
Axes-level → “Draw on this plot”
Figure-level → “Create the whole plot system”

## Hierarchy

relplot() → Figure-level (top controller)  
&nbsp;&nbsp;&nbsp;&nbsp;├── scatterplot() → Axes-level (kind='scatter')  
&nbsp;&nbsp;&nbsp;&nbsp;└── lineplot() → Axes-level (kind='line')  

👉 `relplot()` manages the **figure + layout** and internally calls  
`scatterplot()` or `lineplot()` to draw actual plots.


# 1. Figure-Level Function → `sns.relplot()`

#### Syntax
    sns.relplot(data=df, x=..., y=..., kind='scatter'/'line')


### Parameters

| Parameter | Description |
|----------|-------------|
| data     | Dataset (DataFrame) |
| x, y     | Variables for axes |
| kind     | Type of plot → `'scatter'` or `'line'` |
| hue      | Color grouping (categorical/numeric) |
| size     | Size mapping (numeric variable) |
| style    | Marker/line style grouping |
| col      | Create subplots by column categories |
| row      | Create subplots by row categories |
| col_wrap | Wrap columns into multiple rows |
| palette  | Color palette |
| height   | Height of each subplot |
| aspect   | Width/height ratio |
| facet_kws| Additional subplot settings |

### Role

- Creates **entire figure**
- Handles **multiple plots (faceting)**
- Automatically manages layout

#### Syntax
    sns.scatterplot(data=df, x=..., y=...)

### Parameters

| Parameter | Description |
|----------|-------------|
| data     | Dataset |
| x, y     | Variables |
| hue      | Color grouping |
| size     | Marker size |
| style    | Marker type |
| palette  | Color scheme |
| markers  | Custom marker styles |
| alpha    | Transparency |
| legend   | Show/hide legend |
| ax       | Target axes (for subplots) |


### Role

- Draws **points (scatter)**
- Works on **single plot (axes)**
- Gives **more control**


# 3. Axes-Level Function → `sns.lineplot()`

#### Syntax
    sns.lineplot(data=df, x=..., y=...)


### Parameters

| Parameter | Description |
|----------|-------------|

| legend   | Show/hide legend |
| ax       | Target axes |


### Role

- Draws **lines (trends)**
- Supports **aggregation + confidence intervals**


# Internal Flow

relplot()  
&nbsp;&nbsp;&nbsp;&nbsp;↓ (creates layout & facets)  
&nbsp;&nbsp;&nbsp;&nbsp;calls scatterplot() / lineplot()  
&nbsp;&nbsp;&nbsp;&nbsp;↓  
plots on each subplot  


# Simple Intuition

- relplot() → **manager (creates full plot system)**  
- scatterplot() → **draws points**  
- lineplot() → **draws trends**


# Final Takeaway

- Figure-level → controls **whole figure + multiple plots**  
- Axes-level → controls **actual drawing on a single plot**

### Interpreting Bivariate Histogram & KDE

# 1. Bivariate Histogram (`sns.histplot(x, y)`)

#### What it shows
- Data divided into **2D bins (grid)**
- Each cell shows **count or density of points**

---

## How to Interpret

### 1. Density (Color Intensity)

- Darker / brighter region → **more data points**
- Lighter region → **fewer points**

👉 Shows where data is concentrated


### 2. Pattern Shape

| Pattern        | Meaning                          |
|----------------|----------------------------------|
| Diagonal ↑     | Positive correlation             |
| Diagonal ↓     | Negative correlation             |
| No pattern     | No relationship                  |


### 3. Spread

- Tight cluster → **low variance**
- Wide spread → **high variance**


### 4. Outliers

- Isolated bins far from cluster  
👉 Indicate unusual data points

---

# 2. Bivariate KDE (`sns.kdeplot(x, y)`)

#### What it shows
- **Smooth contour lines** representing density  
- Like a **topographic map**

---

## How to Interpret

### 1. Contour Lines

- Each line = **same density level**
- Inner curves → **higher density**
- Outer curves → **lower density**


### 2. Peaks (Modes)

- Center of contours → **highest concentration**
- Multiple peaks → **multiple clusters**


### 3. Shape of Contours

| Shape        | Meaning                          |
|--------------|----------------------------------|
| Circular     | No correlation                   |
| Elliptical   | Correlation exists               |
| Tilted ellipse ↑ | Positive correlation         |
| Tilted ellipse ↓ | Negative correlation         |


### 4. Distance Between Lines

- Close lines → **sharp change (steep density)**
- Wide gaps → **gradual change**

---

# Histogram vs KDE (Bivariate)

| Feature        | Histogram              | KDE                     |
|---------------|------------------------|--------------------------|
| Representation| Grid (bins)            | Smooth contours          |
| Detail        | Exact counts           | Smoothed distribution    |
| Noise         | More noisy             | Less noisy               |
| Insight       | Raw distribution       | Pattern & trend          |

---

# Combined Interpretation Strategy

When you see both:

1. Use histogram → **see raw data concentration**
2. Use KDE → **understand pattern/relationship**

---

#  Simple Intuition

- Histogram → “where points are packed”  
- KDE → “shape of the distribution”  

👉 Histogram = raw  
👉 KDE = smooth understanding  

---

# Final Takeaway

```text
Bivariate histogram → shows counts
Bivariate KDE → shows smooth density + relationship

# Categorical Scatter Plot

A categorical scatter plot is used to visualize **numerical values across different categories** using scattered points.


### Key Idea

- X-axis → **categorical variable**
- Y-axis → **numerical variable**
- Points → individual observations

#### Syntax

#### Parameters

| Parameter | Description |
|----------|-------------|
| data     | Dataset |
| x, y     | Variables (one categorical, one numeric) |
| hue      | Grouping variable |

### Error Bar

Error bars represent the **uncertainty or variability** in the data around a central value (like mean).


#### Definition

- Shows how much the data **can vary from the estimated value**
- Gives an idea of **confidence or reliability**


| Term          | Meaning                                      |
|---------------|----------------------------------------------|
| Central value | Mean / median                                |
| Error bar     | Range of possible variation                  |
| Spread        | How uncertain the estimate is                |


#### Key Points

- Longer error bar → **more variability / less certainty**  
- Shorter error bar → **less variability / more certainty**  
- Commonly represents:
  - Confidence Interval (CI)  
  - Standard Deviation (SD)  
  - Standard Error (SE)  


#### Key Interpretation

- Overlapping error bars → difference may NOT be significant  
- Non-overlapping → likely significant difference  
- Helps compare **groups statistically**

### Pair Grid (`sns.PairGrid()`)

It is used to create a **grid of subplots for pairwise relationships**, giving **full control over what plots to draw**.


#### Syntax
    g = sns.PairGrid(data=df)
    g.map(func)


| Parameter | Description                                      | Default | Possible Values                  |
|----------|--------------------------------------------------|---------|----------------------------------|
| data     | Dataset                                          | —       | DataFrame                        |
| hue      | Grouping variable (color encoding)               | None    | column name                      |
| vars     | Subset of variables                              | None    | list of columns                  |
| x_vars   | Variables for x-axis                             | None    | list of columns                  |
| y_vars   | Variables for y-axis                             | None    | list of columns                  |
| palette  | Color scheme                                     | None    | seaborn palette                  |
| corner   | Show only lower triangle                         | False   | True, False                      |


#### Key Points

- More **flexible version of `pairplot()`**  
- Requires manual mapping:
  - `map()` → same plot everywhere  
  - `map_diag()` → diagonal plots  
  - `map_upper()` / `map_lower()` → different plots  
- Used for **custom multi-variable visualization**  


#### Key Interpretation

- Each cell → relationship between two variables  
- Diagonal → distribution of individual variables  
- Different plots can be applied to different regions  


#### Key Insight

- `pairplot()` = automatic  
- `PairGrid()` = manual + customizable  

````python
g = sns.PairGrid(data=iris_df,hue='Species',vars=['SepalLengthCm', 'SepalWidthCm'])
g.map(sns.scatterplot)
g.add_legend()
````

````python
g = sns.PairGrid(data=iris_df,hue='Species',vars=['SepalLengthCm', 'SepalWidthCm'])
g.map_diag(sns.scatterplot)
g.map_offdiag(sns.histplot)
g.add_legend()
````

````python
g = sns.PairGrid(data=iris_df,hue='Species',vars=['SepalLengthCm', 'SepalWidthCm'])
g.map_lower(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.add_legend()
````
