### .groupby() 

It is used to group data based on one or more columns and perform aggregate, transformation, or filtering operations.

- **Syntax**

        df.groupby(by=None, axis=0)

| Parameter   | Description                                              | Default | Possible Values                          |
|-------------|----------------------------------------------------------|---------|------------------------------------------|
| by          | Column(s) or key(s) to group by                          | None    | label, list, function, Series            |
| axis        | Axis along which grouping is performed                   | 0       | 0 (rows), 1 (columns)                    |
| level       | Group by a specific level (for MultiIndex)               | None    | int or level name                        |
| as_index    | Return grouped labels as index                           | True    | True, False                              |
| sort        | Sort group keys                                          | True    | True, False                              |
| group_keys  | Include group keys in result index                       | True    | True, False                              |
| observed    | Show only observed values (for categorical data)         | False   | True, False                              |
| dropna      | Exclude NA values from group keys                        | True    | True, False                              |

```python
grouped_genre_data = imdb_data.groupby(by=['Genre'])
grouped_genre_data
```

### Common Functions with `.groupby()`

These functions are typically used after applying `.groupby()`:

---

#### Aggregate Functions (reduce data)

| Function | Description                          |
|----------|--------------------------------------|
| sum()    | Sum of values in each group          |
| mean()   | Average of each group                |
| median() | Median value                         |
| min()    | Minimum value                        |
| max()    | Maximum value                        |
| count()  | Count of non-null values             |
| size()   | Total size of group (including NaN)  |
| std()    | Standard deviation                   |
| var()    | Variance                             |
| agg()    | Apply multiple aggregate functions   |

---

#### Transformation Functions (same shape output)

| Function    | Description                                      |
|-------------|--------------------------------------------------|
| transform() | Apply function and return same shape output      |
| cumcount()  | Number each item in group                        |
| cumsum()    | Cumulative sum within group                      |
| cummax()    | Cumulative maximum                               |
| cummin()    | Cumulative minimum                               |
| rank()      | Rank within each group                           |

---

#### Filter Function

| Function | Description                                  |
|----------|----------------------------------------------|
| filter() | Filter groups based on condition              |

---
