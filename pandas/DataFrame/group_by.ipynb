### .groupby() 

It is used to group data based on one or more columns and perform aggregate, transformation, or filtering operations.

- **Syntax**

        df.groupby(by=None, axis=0)

| Parameter   | Description                                              | Default | Possible Values                          |
|-------------|----------------------------------------------------------|---------|------------------------------------------|
| by          | Column(s) or key(s) to group by                          | None    | label, list, function, Series            |
| axis        | Axis along which grouping is performed                   | 0       | 0 (rows), 1 (columns)                    |
| level       | Group by a specific level (for MultiIndex)               | None    | int or level name                        |
| as_index    | Return grouped labels as index                           | True    | True, False                              |
| sort        | Sort group keys                                          | True    | True, False                              |
| group_keys  | Include group keys in result index                       | True    | True, False                              |
| observed    | Show only observed values (for categorical data)         | False   | True, False                              |
| dropna      | Exclude NA values from group keys                        | True    | True, False                              |

```python
grouped_genre_data = imdb_data.groupby(by=['Genre'])
grouped_genre_data
```

### Common Functions with `.groupby()`

These functions are typically used after applying `.groupby()`:

---

#### Aggregate Functions (reduce data)

| Function | Description                          |
|----------|--------------------------------------|
| sum()    | Sum of values in each group          |
| mean()   | Average of each group                |
| median() | Median value                         |
| min()    | Minimum value                        |
| max()    | Maximum value                        |
| count()  | Count of non-null values             |
| size()   | Total size of group (including NaN)  |
| std()    | Standard deviation                   |
| var()    | Variance                             |
| agg()    | Apply multiple aggregate functions   |

---

#### Transformation Functions (same shape output)

| Function    | Description                                      |
|-------------|--------------------------------------------------|
| transform() | Apply function and return same shape output      |
| cumcount()  | Number each item in group                        |
| cumsum()    | Cumulative sum within group                      |
| cummax()    | Cumulative maximum                               |
| cummin()    | Cumulative minimum                               |
| rank()      | Rank within each group                           |

---

#### Filter Function

| Function | Description                                  |
|----------|----------------------------------------------|
| filter() | Filter groups based on condition              |

---

### .groupby().first() 

It is used to return the **first non-null value** of each column within each group.

- **Syntax**

        df.groupby(by).first()

| Parameter | Description                          | Default | Possible Values |
|-----------|--------------------------------------|---------|-----------------|
| by        | Column(s) to group by                | —       | label, list     |

---

#### Key Points

- Returns the **first non-NaN row** in each group  
- Works column-wise (skips missing values)  
- Commonly used after sorting to get **top record per group**


```python
grouped_data = imdb_data.groupby(by='Genre')
```

```python
grouped_data.first() #gives the first non-zero data of each grouped feature
```

### .groupby().last() 

It is used to return the **last non-null value** of each column within each group.

---

#### Key Points

- Returns the **last non-NaN row** in each group  
- Works column-wise (skips missing values)  
- Often used after sorting to get the **latest record per group**

### .groupby().nth() 

It is used to return the **nth row** from each group.

- **Syntax**

        df.groupby(by).nth(n)

| Parameter | Description                          | Default | Possible Values            |
|-----------|--------------------------------------|---------|----------------------------|
| by        | Column(s) to group by                | —       | label, list                |
| n         | Position of row to select            | —       | int, list of ints          |
| dropna    | Exclude NaN values before selection  | None    | 'any', 'all', None         |

---

#### Key Points

- Returns row(s) at a specific position from each group  
- Unlike `.first()` / `.last()`, it **does not skip NaN by default**  
- Can select multiple positions using a list  


```python
grouped_data.nth(6) #extract 7th row of each group formed
```

```python
grouped_data.nth([4,8,9]) # extract 5th, 9th and 10th row of each group formed
```

### .groupby().get_group() 

It is used to retrieve a specific group from a GroupBy object.

- **Syntax**

        df.groupby(by).get_group(name)

| Parameter | Description                          | Default | Possible Values |
|-----------|--------------------------------------|---------|-----------------|
| by        | Column(s) to group by                | —       | label, list     |
| name      | Group key to retrieve                | —       | value, tuple    |

---

#### Key Points

- Returns all rows corresponding to a specific group  
- Useful for inspecting or operating on a single group  
- `name` should match one of the group keys  

```python
grouped_data.get_group('Mystery') #get all data which is present in Mystery group
```

### .groupby().sample() 

It is used to randomly sample rows from each group.

- **Syntax**

        df.groupby(by).sample(n=None, frac=None)

| Parameter    | Description                                              | Default | Possible Values                          |
|--------------|----------------------------------------------------------|---------|------------------------------------------|
| by           | Column(s) to group by                                    | —       | label, list                              |
| n            | Number of items to return per group                      | None    | int                                      |
| frac         | Fraction of items to return per group                    | None    | float (0–1)                              |
| replace      | Allow sampling with replacement                          | False   | True, False                              |
| random_state | Seed for reproducibility                                 | None    | int                                      |

---

#### Key Points

- Performs **random sampling within each group**  
- Use `n` for fixed number or `frac` for percentage  
- Helps in **balanced sampling** from grouped data  

```python
grouped_data.sample() #get a random data from each group
```

### .groupby().agg() 

It is used to apply one or multiple **aggregate functions** to grouped data.

- **Syntax**

        df.groupby(by).agg(func)

| Parameter | Description                                              | Default | Possible Values                          |
|-----------|----------------------------------------------------------|---------|------------------------------------------|
| by        | Column(s) to group by                                    | —       | label, list                              |
| func      | Function(s) to apply                                     | —       | function, string, list, dict             |
| axis      | Axis along which aggregation is performed                | 0       | 0 (rows), 1 (columns)                    |

---

#### Key Points

- Can apply **single or multiple aggregations**  
- Supports:
  - list of functions → `['mean', 'sum']`  
  - dictionary mapping → `{'col1': 'mean', 'col2': 'sum'}`  
- Returns aggregated result per group  

```python
grouped_data.agg([min, max]) # getting min & max for each column
```

```python
#custom aggregate function
grouped_data.agg(
    {
        'Runtime': ['mean','max','min'],
        'IMDB_Rating':['mean','max','min'],
        'Gross':'sum'
    }
)
```

### Looping in Grouped Data (`.groupby()`)

It is used to iterate over each group in a GroupBy object to perform custom operations.

- **Syntax**

        for name, group in df.groupby(by):
            # operations on group

| Parameter | Description                          | Default | Possible Values |
|-----------|--------------------------------------|---------|-----------------|
| by        | Column(s) to group by                | —       | label, list     |
| name      | Group key (unique value of group)    | —       | varies          |
| group     | Subset of DataFrame for that group   | —       | DataFrame       |

---

#### Key Points

- Iterates over **(group_name, group_data)** pairs  
- `group` is a **DataFrame subset** corresponding to that group  
- Useful for **custom logic not covered by built-in functions**  

```python
for group, data in grouped_data:
    print(group)
    print(' ')
    print(data)
```

### .groupby().apply() 

It is used to apply a custom function to each group and combine the results.

- **Syntax**

        df.groupby(by).apply(func)

| Parameter | Description                                              | Default | Possible Values                          |
|-----------|----------------------------------------------------------|---------|------------------------------------------|
| by        | Column(s) to group by                                    | —       | label, list                              |
| func      | Function to apply to each group                          | —       | function, lambda                         |
| axis      | Axis along which function is applied                     | 0       | 0 (rows), 1 (columns)                    |

---

#### Key Points

- Applies **custom logic** to each group  
- More flexible than `.agg()` and `.transform()`  
- Can return **DataFrame, Series, or scalar**  
- Slower compared to built-in aggregations

### Find ranking of each movie in the group according to IMDB score

### .rank() 

It is used to assign ranks to values in a Series or DataFrame.

- **Syntax**

        df.rank(axis=0, method='average', ascending=True)

| Parameter   | Description                                              | Default     | Possible Values                                      |
|-------------|----------------------------------------------------------|-------------|------------------------------------------------------|
| axis        | Axis along which ranking is applied                      | 0           | 0 (index), 1 (columns)                               |
| method      | Method to assign ranks in case of ties                   | 'average'   | 'average', 'min', 'max', 'first', 'dense'            |
| ascending   | Rank order                                               | True        | True, False                                          |
| na_option   | How to handle NaN values                                 | 'keep'      | 'keep', 'top', 'bottom'                              |
| pct         | Return ranks as percentage                               | False       | True, False                                          |
| numeric_only| Include only numeric data (DataFrame)                    | None        | True, False                                          |

---

#### Key Points

- Assigns ranks to values (1 = smallest by default)  
- Handles ties using different methods  
- Can rank across rows or columns  


```python
def ranking_in_group(group):
    group['ranking_in_group'] = group['IMDB_Rating'].rank(method = 'min', ascending = False)
    return group
```

```python
grouped_data.apply(ranking_in_group)
```

### Find the normalized IMDB rating group wise

```python
def normalize(group):
    group['normalized_rating'] = (group['IMDB_Rating']-group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()-group['IMDB_Rating'].min())
    return group
```

### Create a function that can return the highest score of any batsman

```python
def highest_score(name):
    player_data = delivery_data[delivery_data['batsman'] == name]
    player_match_detail = player_data.groupby(by=['match_id'], dropna = False)
    return player_match_detail['batsman_runs'].sum().max()
```

```python
highest_score('V Kohli')
```

```python
highest_score('CH Gayle')
```
