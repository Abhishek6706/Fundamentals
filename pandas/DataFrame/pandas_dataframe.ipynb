### Dataframe core components

**Two-Dimensional**: It organizes data into a grid of rows and columns (an $X$ and $Y$ axis).

**Heterogeneous**: Unlike a NumPy array, a single DataFrame can store a string in Column A, an integer in Column B, and a boolean in Column C.

**Labeled Axes**: It uses an Index for rows and Columns for vertical headers, allowing you to access data by name (e.g., df['Price']) rather than just by position.

### .rename()

The .rename() method is used to change the labels of columns or the index in a DataFrame or Series. 

It is highly flexible because you can rename specific labels without having to redefine the entire list of headers.

1. **Renaming Columns**

To change column names, you pass a dictionary to the columns parameter where the key is the old name and the value is the new name.
- **Syntax**: 
            
        df.rename(columns={'old_name': 'new_name'})

- **Output**: 

    A new DataFrame with the updated headers (unless inplace=True is used).
    
- **Common In**: 

    Cleaning up inconsistently named data, removing spaces from headers, or preparing data for a SQL export.
    
2. **Renaming the Index**

Similarly, you can change specific row labels by passing a dictionary to the index parameter.

- **Syntax**: 

        df.rename(index={0: 'Match_1', 1: 'Match_2'})
        
- **Output**: 

A new DataFrame with the updated row labels.

- **Common In**: 

Changing date formats in an index or mapping numeric IDs to descriptive names.

3. **Using Functions (Mappers)**

Instead of a dictionary, you can pass a function (like str.upper or str.lower) to transform all labels at once.

- **Syntax**: 

        df.rename(columns=str.upper)
        
- **Output**: A DataFrame where every column name has been modified by the function.

- **Common In**: Standardizing headers to all lowercase or uppercase for easier coding.

### .sum()
The .sum() method calculates the total of all values in a Series or along a specified axis in a DataFrame. By default, it skips missing values (NaN) and returns the arithmetic sum of the data.

- **Syntax**: 

        df.sum(axis=0, skipna=True, numeric_only=False)

    *changing the axis to 1 (or 'columns') calculates the sum horizontally across each row*

- **Output**: A single number (for a Series) or a Series of sums (for a DataFrame).

- **Common In**: Calculating total revenue, total distance traveled, or aggregating scores for a specific time period.

### .set_index()

The .set_index() method is used to promote one or more existing columns in a DataFrame to become the Index (the row labels). 

This is a fundamental step in data preparation, especially when you want to move away from the default $0, 1, 2, 3...$ numbering.

- **Syntax**: 
        df.set_index(keys, drop=True, append=False, inplace=False)
        
- **Output**: A new DataFrame where the specified column is now the bolded "label" on the left.

- **Common In**: Converting "Date" columns for time-series analysis or using "ID" columns for easier data alignment.

```python
stock_data_modified = stocks_data.set_index(keys='Stock Name')
stock_data_modified
```

### Selecting rows from the dataframe

Done by two methods:
1. **iloc** : searches using index positions
2. **loc** : searches using index labels

In Pandas, .loc (Label-based selection) and .iloc (Integer-based selection) are the primary ways to access data in a DataFrame. While they look like functions, they are actually indexers, which is why they use square brackets [] instead of parentheses ().

### .loc (Label-based)

.loc is used to access a group of rows and columns by labels or a boolean array. If you have set a specific column (like 'Match_ID') as your index, you use those IDs to find data.

- **Syntax**: 

        df.loc[row_label, column_label]

- **Slicing**: 
Unlike standard Python slicing, .loc is inclusive of the stop value.

*df.loc[100:105] will give you rows 100, 101, 102, 103, 104, and 105.*

- **Common In**: Selecting specific categories, filtering by date, or updating values based on a condition.

### .iloc (Integer-based)
.iloc is purely positional. It doesn't care what the labels are; it only cares about where the data sits in the table ($0, 1, 2...$).

- **Syntax**: 

        df.iloc[row_position, column_position]
    
- **Slicing**: This follows standard Python rules: exclusive of the stop value.df.iloc[0:5] gives you positions 0, 1, 2, 3, and 4.
- **Common In**: Taking the "first 10 rows," splitting data into training/testing sets, or selecting columns by their index number (e.g., "give me the last column").

### Write a function that can return the track record of 2 teams against each other

```python
team1 = 'Kings XI Punjab'
team2 = 'Royal Challengers Bangalore'
```

```python
criteria1 = ipl_data['Team1'].isin([team1, team2])
criteria2 = ipl_data['Team2'].isin([team1, team2])

criteria = criteria1 & criteria2
criteria
```

```python
winning_matches = ipl_data[criteria].WinningTeam.value_counts()[0]
winning_matches
```

```python
winning_matches.index[0]
```

```python
def track_record(team1, team2):
    criteria1 = ipl_data['Team1'].isin([team1,team2])
    criteria2 = ipl_data['Team2'].isin([team1,team2])

    criteria = criteria1 & criteria2

    t1_name = ipl_data[criteria].WinningTeam.value_counts().index[0]
    t1_winning_game = ipl_data[criteria].WinningTeam.value_counts()[0]

    t2_name = ipl_data[criteria].WinningTeam.value_counts().index[1]
    t2_winning_game = ipl_data[criteria].WinningTeam.value_counts()[1]

    t1_winning_ratio = (t1_winning_game/(t1_winning_game+t2_winning_game))*100
    t2_winning_ratio = (t2_winning_game/(t1_winning_game+t2_winning_game))*100

    print('Team', t1_name, 'has the win ratio of', t1_winning_ratio, 'against', t2_name)
    print('Team', t2_name, 'has the win ratio of', t2_winning_ratio, 'against', t1_name)

    # return t1_name,t2_name,t1_winning_ratio,t2_winning_ratio

```

```python
team1 = 'Kings XI Punjab'
team2 = 'Royal Challengers Bangalore'
track_record(team1, team2)
```
