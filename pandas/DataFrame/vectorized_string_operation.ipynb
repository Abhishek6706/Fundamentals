### `.str` (Vectorized String Operations)

It is used to perform **string operations on entire Pandas Series efficiently** (without using loops).


#### Syntax
    df['column'].str.method()


| Method        | Description                                      | Example                          |
|---------------|--------------------------------------------------|----------------------------------|
| lower()       | Convert to lowercase                             | str.lower()                      |
| upper()       | Convert to uppercase                             | str.upper()                      |
| title()       | Convert to title case                            | str.title()                      |
| strip()       | Remove leading/trailing spaces                   | str.strip()                      |
| lstrip()      | Remove left spaces                               | str.lstrip()                     |
| rstrip()      | Remove right spaces                              | str.rstrip()                     |
| replace()     | Replace substring                                | str.replace('a','b')             |
| contains()    | Check substring presence                         | str.contains('abc')              |
| startswith()  | Check prefix                                     | str.startswith('A')              |
| endswith()    | Check suffix                                     | str.endswith('z')                |
| len()         | Length of string                                 | str.len()                        |
| split()       | Split string                                     | str.split(',')                   |
| get()         | Extract element from split                       | str.split(',').str.get(0)        |
| extract()     | Extract using regex                              | str.extract('(\d+)')             |
| find()        | Find substring position                          | str.find('x')                    |


```python
import pandas as pd
```

```python
titanic_data = pd.read_csv('titanic.csv')
```

```python
titanic_data.head()
```

### Make 3 columns from Name, i.e., Salutaion, First Name and Last Name

```python
titanic_data['Name']
```

```python
titanic_data['Name'].str.split(',') #splitted first on the basis of ',' because the first one is the last name
```

```python
titanic_data['Last_Name'] = titanic_data['Name'].str.split(',').str.get(0)
```

```python
titanic_data['Last_Name']
```

```python
titanic_data['Name'].str.split(',').str.get(1).str.split('.') #splitted on the basis of '.', because the first one is the salutation
```

```python
titanic_data['Salutation'] = titanic_data['Name'].str.split(',').str.get(1).str.split('.').str.get(0)
```

```python
titanic_data['Salutation']
```

```python
titanic_data['First_Name'] = titanic_data['Name'].str.split(',').str.get(1).str.split('.').str.get(1)
```

```python
titanic_data['First_Name']
```

```python
titanic_data[['Salutation', 'First_Name', 'Last_Name']]
```

### Find out the person whose First_Name have 'x' 

```python
titanic_data[titanic_data['First_Name'].str.contains('x')]
```

### Find the person whose last name starts with 'P'

```python
titanic_data[titanic_data['Last_Name'].str.startswith('P')]
```
