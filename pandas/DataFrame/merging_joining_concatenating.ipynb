### pd.concat() 

It is used to concatenate (combine) multiple Series or DataFrames along a particular axis.

- **Syntax**

        pd.concat(objs, axis=0)

| Parameter        | Description                                              | Default | Possible Values                          |
|------------------|----------------------------------------------------------|---------|------------------------------------------|
| objs             | List or sequence of Series/DataFrames to concatenate     | —       | list, tuple                              |
| axis             | Axis along which to concatenate                          | 0       | 0 (rows), 1 (columns)                    |
| join             | How to handle indexes on other axis                      | 'outer' | 'inner', 'outer'                         |
| ignore_index     | Reset index in the result                                | False   | True, False                              |
| keys             | Create hierarchical index                                | None    | list                                     |
| names            | Names for hierarchical index                             | None    | list                                     |
| verify_integrity | Check for duplicate indexes                              | False   | True, False                              |
| sort             | Sort non-concatenation axis                              | False   | True, False                              |

```python
all_registered_student = pd.concat([registration_month_1_df, registration_month_2_df], axis=0, ignore_index=True) #vertical merging
```

```python
all_registered_student
```

```python
registered_students = pd.concat([registration_month_1_df,registration_month_2_df], ignore_index=True, axis=1) # horizontal merging
```

```python
registered_students
```

### MultiIndex DataFrame (Hierarchical Indexing)

A **MultiIndex DataFrame** is a Pandas DataFrame that has **multiple levels of indexing** on rows and/or columns, allowing you to represent higher-dimensional data in a 2D structure.

---

#### 🔹 Key Concepts

- Also called **hierarchical indexing**  
- Enables grouping and selection across multiple levels  
- Can exist on:
  - Rows (`index`)
  - Columns (`columns`)
- Useful for **grouped, pivoted, or structured data**

```python
multi_df = pd.concat([registration_month_1_df,registration_month_2_df], keys=['Nov', 'Dec'])
```

```python
multi_df
```

```python
# to extract a specific row

multi_df.loc[('Nov', 5)]
```

### Inner Join (Pandas)

It is used to combine two DataFrames by keeping only the **common (matching) values** from both tables.

---

#### Syntax

        pd.merge(df1, df2, how='inner', on=key)

| Parameter | Description                                      | Default | Possible Values                          |
|-----------|--------------------------------------------------|---------|------------------------------------------|
| df1       | Left DataFrame                                   | —       | DataFrame                                |
| df2       | Right DataFrame                                  | —       | DataFrame                                |
| how       | Type of join                                     | 'inner' | 'inner', 'left', 'right', 'outer'        |
| on        | Column(s) to join on                             | None    | column label, list                       |
| left_on   | Column(s) from left DataFrame                    | None    | column label, list                       |
| right_on  | Column(s) from right DataFrame                   | None    | column label, list                       |
| left_index| Use index from left DataFrame                    | False   | True, False                              |
| right_index| Use index from right DataFrame                  | False   | True, False                              |
| suffixes  | Suffixes for overlapping column names            | ('_x','_y') | tuple                                |

---

#### Key Points

- Keeps only **intersection of keys**  
- Drops non-matching rows from both DataFrames  
- Most commonly used join in real-world datasets  


```python
inner_join_df = pd.merge(all_registered_student, student_df, how='inner', on='student_id')
```

```python
inner_join_df.sort_values(by=['student_id']).reset_index()
```

### Left Join (Pandas)

It is used to combine two DataFrames by keeping **all rows from the left DataFrame** and matching rows from the right DataFrame.

---
#### Key Points

- Keeps **all rows from left DataFrame**  
- Non-matching rows from right → filled with **NaN**  
- Useful when left dataset is the **primary/reference data**

```python
left_df = pd.merge(all_registered_student, student_df, how='left', on='student_id')
```

```python
left_df
```

### Right Join (Pandas)

It is used to combine two DataFrames by keeping **all rows from the right DataFrame** and matching rows from the left DataFrame.

#### Key Points

- Keeps **all rows from right DataFrame**  
- Non-matching rows from left → filled with **NaN**  
- Useful when right dataset is the **primary/reference data**  

```python
pd.merge(course_df, all_registered_student, how= 'right', on= 'course_id')
```

### Outer Join (Pandas)

It is used to combine two DataFrames by keeping **all rows from both DataFrames**, filling non-matching values with NaN.

#### Key Points

- Keeps **all rows from both DataFrames**  
- Non-matching values are filled with **NaN**  
- Also known as **full outer join**  
- Useful when you want **complete data without losing anything**  


```python
pd.merge(course_df, all_registered_student, how='outer', on= 'course_id')
```

##### Different way to solve the same problem

### np.intersect1d() 

It is used to find the **common (intersection) elements** between two arrays.

- **Syntax**

        np.intersect1d(ar1, ar2)

| Parameter | Description                              | Default | Possible Values        |
|-----------|------------------------------------------|---------|------------------------|
| ar1       | First input array                        | —       | array-like             |
| ar2       | Second input array                       | —       | array-like             |
| assume_unique | Assume inputs are unique (faster)   | False   | True, False            |
| return_indices | Return indices of common elements   | False   | True, False            |

---

#### Key Points

- Returns **sorted unique common elements**  
- Works with NumPy arrays or array-like inputs  
- Removes duplicates automatically  
- Useful for **set-like operations**  

```python
registration_month_1_df['student_id']
```

```python
registration_month_2_df['student_id']
```

```python
common_students = np.intersect1d(registration_month_1_df['student_id'],registration_month_2_df['student_id'])
```

```python
#to get the name of each student who enrolled in both months

student_df[student_df['student_id'].isin(common_students)]['name']
```

#### Different way to solve the same problem

### np.setdiff1d() 

It is used to find the **elements present in the first array but not in the second array** (set difference).

- **Syntax**

        np.setdiff1d(ar1, ar2)

| Parameter       | Description                              | Default | Possible Values        |
|-----------------|------------------------------------------|---------|------------------------|
| ar1             | First input array                        | —       | array-like             |
| ar2             | Second input array                       | —       | array-like             |
| assume_unique   | Assume inputs are unique (faster)        | False   | True, False            |

---

#### Key Points

- Returns **sorted unique values** in `ar1` that are not in `ar2`  
- Removes duplicates automatically  
- Performs a **set-like subtraction** operation  

```python
not_enrolled_courses_index = np.setdiff1d(course_df['course_id'], all_registered_student['course_id'])
```

```python
not_enrolled_courses_index
```

```python
course_df['course_id'].isin(not_enrolled_courses_index)
```

### Self Join (Pandas)

It is used to join a DataFrame with **itself** to relate rows within the same dataset.

#### Syntax

        pd.merge(df, df, how='inner', on=key)

| Parameter | Description                                      | Default | Possible Values                          |
|-----------|--------------------------------------------------|---------|------------------------------------------|
| df        | DataFrame to join with itself                    | —       | DataFrame                                |
| how       | Type of join                                     | 'inner' | 'inner', 'left', 'right', 'outer'        |
| on        | Column(s) to join on                             | None    | column label, list                       |
| left_on   | Column(s) from left side                         | None    | column label, list                       |
| right_on  | Column(s) from right side                        | None    | column label, list                       |
| suffixes  | Suffixes for overlapping column names            | ('_x','_y') | tuple                                |

---

#### Key Points

- Joins a table **with itself**  
- Useful for:
  - hierarchical relationships (e.g., employee-manager)  
  - comparing rows within same dataset  
- Requires **aliases (suffixes)** to differentiate columns  

```python
pd.merge(student_df, student_df, how='inner', left_on=['partner'], right_on=['student_id'])
```

```python
pd.merge(student_df, student_df, how='inner', left_on=['partner'], right_on=['student_id'])[['name_x', 'name_y']]
```
