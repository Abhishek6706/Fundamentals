The .sort_values() method is used to rearrange your rows based on the data in one or more columns. 
By default, it sorts in ascending order (lowest to highest, or A to Z).

1. **Basic Sorting (Series)**

If you have a single Series, calling .sort_values() simply reorders the numbers.

- **Syntax**: 

        s.sort_values(ascending=True)

2. **Sorting a DataFrame**

When working with a full table, we must tell Pandas which column to use as the key for sorting using the by parameter.

- **Syntax**: 

        df.sort_values(by='column_name')
        
3. **Sorting by Multiple Columns**

We can sort by a hierarchy of columns by passing a list. This is useful for "tie-breaking."

- **Syntax**: 

        s.sort_values(by=['column1', 'column2'], ascending=[True, False])

- **Logic**: This sorts the data by Column1 first (lowest to highest), and within each Column1 value, it sorts the Column2 value by the highest to lowest.

```python
ipl_data.sample()
```

```python
ipl_data.sort_values(by='Margin', ascending=False)
```

```python
movies_data.sample()
```

```python
movies_data.sort_values(by=['original_title','year_of_release'], ascending=[True, False])
```

### .rank()

The .rank() method provides a numerical rank ($1, 2, 3...$) for each element in a Series.

By default, .rank() assigns the rank $1$ to the smallest value (ascending).

- **Syntax**: s.rank(ascending=True, method='average')

| Method  | Behavior                                      | Example (Scores: 50, 50, 80) |
|---------|----------------------------------------------|------------------------------|
| average | Assigns the average rank to all ties         | 1.5, 1.5, 3.0                |
| min     | Assigns the lowest rank (competition rank)   | 1, 1, 3                      |
| max     | Assigns the highest rank                    | 2, 2, 3                      |
| first   | Ranks based on order of appearance          | 1, 2, 3                      |
| dense   | Like min, but no gaps in ranking            | 1, 1, 2                      |

```python
batsman_data.head(1)
```

```python
batsman_data['Rank'] = batsman_data['batsman_run'].rank(ascending=False)
batsman_data
```

```python
batsman_data.sort_values(by='Rank', ascending= True)
```

### .sort_index() 

It is used to sort a DataFrame or Series based on its index (row labels or column labels) instead of values.

- **Syntax**

        df.sort_index(axis=0, ascending=True, inplace=False)

| Parameter   | Description                                              | Default | Possible Values              |
|-------------|----------------------------------------------------------|---------|------------------------------|
| axis        | Axis to sort along                                       | 0       | 0 (rows), 1 (columns)        |
| level       | Sort by a specific level (for MultiIndex)                | None    | int or level name            |
| ascending   | Sort order                                               | True    | True, False                  |
| inplace     | Modify original object or return new                     | False   | True, False                  |
| kind        | Sorting algorithm                                        | 'quicksort' | 'quicksort', 'mergesort', 'heapsort', 'stable' |
| na_position | Position of NaN values                                   | 'last'  | 'first', 'last'              |
| sort_remaining | Sort remaining levels (MultiIndex)                   | True    | True, False                  |
