### .head()
This method returns the first $n$ rows of your DataFrame or Series. 
- **Syntax**: 

        df.head(n=5)

- **Output**: A new object containing the top $n$ rows (default is 5).
- **Common In**: Checking if a file loaded with the correct column names and identifying the initial structure of the data.

### .tail()
This method returns the last $n$ rows of your DataFrame or Series.

- **Syntax**: 

        df.tail(n=5)

- **Output**: A new object containing the bottom $n$ rows (default is 5).

### .sample()
This method returns a random selection of $n$ rows from your DataFrame or Series. It is much more reliable than head() or tail() for finding hidden patterns or data quality issues.

- **Syntax**: 

        df.sample(n=1, frac=None, random_state=None)

- **Output**: A new object with $n$ randomly chosen rows (default is 1).

### .value_counts()
The .value_counts() method is used to count the occurrences of unique values in a Series or DataFrame column. It returns a new Series where the index contains the unique values and the data contains their respective frequencies, sorted in descending order by default.

- **Syntax**: 

        s.value_counts(ascending=False, normalize=False, bins=None, dropna=True)

*When the normalize parameter is set to True, the function returns the relative frequencies (proportions) instead of the raw counts.*

*Instead of counting individual numbers, it groups them into $n$ equal-width intervals (half-open bins).*

- **Output**: A Series indexed by the unique values from the original data, with counts as the values.

**Common In**: Categorical data analysis (e.g., counting wins per team) or identifying the most frequent observations in a dataset.







```python
kohli_runs_series.value_counts()
```

```python
bollywood_series.value_counts()
```

```python
subscriber_series.value_counts()
```

### .sort_values()
The .sort_values() method is used to sort a Pandas Series or DataFrame based on its values. By default, it sorts in ascending order (lowest to highest) and creates a new object rather than modifying the original one.

- **Syntax**: s.sort_values(ascending=True, inplace=False, na_position='last')

- **Output**: A new Series or DataFrame with the rows reordered according to the data values.

- **Common In**: Finding the "Top 10" records, identifying the oldest or newest dates, and preparing data for visualization.

```python
subscriber_series.sort_values()
```

```python
# finding max

subscriber_series.sort_values().tail(1).values[0]
```

### .sort_index()
The .sort_index() method is used to sort a Series or DataFrame by its labels (the index) rather than its actual data values. This is particularly useful for returning a dataset to its original chronological order after it has been shuffled or sorted by values.

- **Syntax**: 
        
        s.sort_index(ascending=True, inplace=False, na_position='last')

- **Output**: A new Series or DataFrame where the rows are ordered by the index labels.
- **Common In**: Restoring time-series data, organizing value_counts() results numerically, and aligning different datasets before merging.

```python
kohli_runs_series.sort_index()
```

```python
bollywood_series.sort_index()
```

### .describe()

The .describe() method is one of the most powerful tools in Pandas for Exploratory Data Analysis (EDA). It generates a summary of descriptive statistics that provide a quick "bird's-eye view" of your dataset’s distribution, central tendency, and dispersion.


#### For Numerical Data
When used on a Series of numbers (like Virat Kohli's scores), it automatically calculates the following 8 values:

 - **count**: Number of non-null (non-NaN) entries.

- **mean**: The average score.

- **std**: Standard deviation (how much his scores vary from the average).

- **min**: The lowest score (e.g., 0).

- **25%**: The 1st quartile (25% of his scores are below this value).

- **50%**: The median (half his scores are above this, half below).

- **75%**: The 3rd quartile (75% of his scores are below this value).

- **max**: The highest score (e.g., 113).

- **Syntax**: s.describe()

- **Common In**: Initial data inspection to find outliers or understand the "spread" of the data.

#### For Categorical Data
If your Series contains text (e.g., Team names or Venues), .describe() changes its output to show:

- **count**: Total non-null entries.

- **unique**: Number of distinct categories (e.g., how many different teams he played against).

- **top**: The most frequent category (the "Mode").

- **freq**: How many times that top category appeared.

- **Syntax**: df['column_name'].describe()

- **Common In**: Summarizing survey results, city lists, or categorical labels.
