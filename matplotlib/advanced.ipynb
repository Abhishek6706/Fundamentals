### `plt.figure()`

It is used to **create a new figure (canvas)** for plotting.


#### Syntax
    plt.figure(figsize=None)


| Parameter | Description                                      | Default | Possible Values                  |
|----------|--------------------------------------------------|---------|----------------------------------|
| figsize  | Size of the figure (width, height in inches)     | None    | (width, height)                  |
| dpi      | Resolution of the figure                         | None    | int (e.g., 100, 300)             |
| facecolor| Background color of figure                       | None    | color                            |
| edgecolor| Border color of figure                           | None    | color                            |


#### What it does

- Creates a **blank canvas**  
- All plots are drawn inside this figure  
- Useful when creating **multiple plots separately**


#### Key Points

- Controls **size and resolution** of plot  
- Helps manage **multiple figures**  
- Should be called **before plotting**  
- Commonly used with `figsize`  


#### Important Note

- If not used → matplotlib creates a default figure automatically  
- Use it when you need **custom control**


#### Key Insight

- `figure()` = canvas  
- `plot()` = drawing on canvas  


````python
plt.figure(figsize=(10,8))
plt.scatter(iris_df['SepalLengthCm'], iris_df['PetalLengthCm'], c=iris_df['Species'].map({'Iris-setosa':'red', 'Iris-versicolor':'green', 'Iris-virginica':'black'}))
plt.title('Relation between Sepal Length and Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
````

### Annotation (`plt.annotate()`)

It is used to **add text with arrows or labels to specific points** in a plot.


#### Syntax
    plt.annotate(text, xy, xytext=None)


| Parameter | Description                                      | Default | Possible Values                  |
|----------|--------------------------------------------------|---------|----------------------------------|
| text     | Text to display                                  | —       | str                              |
| xy       | Point to annotate (x, y)                         | —       | tuple                            |
| xytext   | Position of text                                 | None    | tuple                            |
| arrowprops | Arrow properties                               | None    | dict                             |
| fontsize | Size of text                                     | None    | int                              |
| color    | Color of text                                    | None    | color                            |
| ha       | Horizontal alignment                             | None    | 'left', 'center', 'right'        |
| va       | Vertical alignment                               | None    | 'top', 'center', 'bottom'        |


#### Key Concepts

- `xy` → actual data point  
- `xytext` → position of text  
- Arrow connects **text → data point**


#### Arrow Properties (`arrowprops`)

| Key        | Description                  |
|------------|------------------------------|
| arrowstyle | Style of arrow               |
| color      | Arrow color                  |
| lw         | Line width                   |


#### Key Points

- Used to **highlight important points**  
- Improves **readability and storytelling**  
- Common in **EDA and presentations**  
- Works well with scatter and line plots  


#### Key Insight

- Annotation = **explanation on plot**  
- Helps viewers understand **important insights quickly**

### `plt.text()`

It is used to **add simple text at a specific location on the plot**.


#### Syntax
    plt.text(x, y, s)


| Parameter | Description                                      | Default | Possible Values                  |
|----------|--------------------------------------------------|---------|----------------------------------|
| x        | X-coordinate of text                             | —       | int, float                       |
| y        | Y-coordinate of text                             | —       | int, float                       |
| s        | Text string to display                           | —       | str                              |
| fontsize | Size of text                                     | None    | int                              |
| color    | Color of text                                    | None    | color                            |
| ha       | Horizontal alignment                             | None    | 'left', 'center', 'right'        |
| va       | Vertical alignment                               | None    | 'top', 'center', 'bottom'        |
| rotation | Rotate text                                      | 0       | int (degrees)                    |


#### Key Points

- Places **text directly at (x, y)**  
- Does **not include arrows** (unlike `annotate()`)  
- Used for **labels, values, notes on plot**  
- Simple and lightweight  


#### `plt.text()` vs `plt.annotate()`

| Feature        | `plt.text()`        | `plt.annotate()`        |
|---------------|---------------------|-------------------------|
| Arrow support | ❌ No               | ✅ Yes                  |
| Simplicity    | ✅ Simple           | ❌ More flexible        |
| Use case      | Labels/values       | Highlight points        |


#### Key Insight

- `text()` = place text  
- `annotate()` = explain with arrow

### `plt.axhline()` and `plt.axvline()`

They are used to draw **reference lines** across the plot.

### `plt.subplots()`

It is used to create **multiple plots (axes) in a single figure**.


#### Syntax
    fig, ax = plt.subplots(nrows=1, ncols=1)


| Parameter | Description                                      | Default | Possible Values                  |
|----------|--------------------------------------------------|---------|----------------------------------|
| nrows    | Number of rows of subplots                       | 1       | int                              |
| ncols    | Number of columns of subplots                    | 1       | int                              |
| figsize  | Size of figure (width, height)                   | None    | tuple                            |
| sharex   | Share x-axis across plots                        | False   | True, False                      |
| sharey   | Share y-axis across plots                        | False   | True, False                      |

#### What it returns

- `fig` → Figure (entire canvas)  
- `ax` → Axes (individual plot area)  


#### Types of `ax`

| Case                     | Type of `ax`        |
|--------------------------|---------------------|
| 1 subplot                | Single Axes object  |
| Multiple subplots        | Array of Axes       |


#### Accessing Subplots

| Layout        | Access Method        |
|--------------|----------------------|
| 1 row, 2 cols| `ax[0]`, `ax[1]`     |
| 2x2 grid     | `ax[0,0]`, `ax[1,1]` |


#### Key Points

- Used for **multiple visualizations in one figure**  
- Each subplot is an independent plot  
- Helps in **comparison and analysis**  
- More flexible than `plt.subplot()`  


#### Key Insight

- `figure` = whole canvas  
- `axes` = individual plots inside it  


### Plot 4 figures in a canva where 2 fig shows the correlation between 'runs' & 'strike_rate' and 'avg' & 'strike_rate', and 2 fig is the histogram of 'avg' and 'strike_rate'

````python
fig, ax = plt.subplots(nrows= 2, ncols= 2, figsize=(15,10))

ax[0,0].scatter(batsman_record_df['avg'], batsman_record_df['strike_rate'])
ax[0,0].set_xlabel('Avg')
ax[0,0].set_ylabel('Strike Rate')
ax[0,0].set_title('Avg vs Strike Rate')

ax[0,1].scatter(batsman_record_df['runs'], batsman_record_df['strike_rate'], color='red')
ax[0,1].set_xlabel('Runs')
ax[0,1].set_ylabel('Strike Rate')
ax[0,1].set_title('Runs vs Strike Rate')

ax[1,0].hist(batsman_record_df['runs'], bins = [0,1000,2000,3000,4000,5000,6000,7000], log=True)
ax[1,0].set_xlabel('Runs')
ax[1,0].set_ylabel('No. of Players')
ax[1,0].set_title('Runs by No. of player')

ax[1,1].hist(batsman_record_df['strike_rate'], bins = [0,50,100,150,200,250,300], log=True)
ax[1,1].set_xlabel('Strike Rate')
ax[1,1].set_ylabel('No. of Players')
ax[1,1].set_title('Strike Rate of Players Distribution')

fig.show()
````

### 3D Scatter Plot

It is created using **Matplotlib’s 3D projection (`projection='3d'`)**.


#### Syntax
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x, y, z)


| Component            | Description                              |
|---------------------|------------------------------------------|
| plt.figure()        | Creates figure (canvas)                  |
| add_subplot(111)    | Adds a subplot (1x1 grid, position 1)    |
| projection='3d'     | Enables 3D plotting                      |
| ax.scatter()        | Plots points in 3D space                 |

#### Parameters in `ax.scatter()`

| Parameter | Description                          |
|----------|--------------------------------------|
| x        | X-axis values                        |
| y        | Y-axis values                        |
| z        | Z-axis values                        |
| c        | Color of points                      |
| s        | Size of points                       |
| marker   | Shape of points                      |
| cmap     | Colormap                             |
| alpha    | Transparency                         |

````python
fig = plt.figure(figsize=(15,7))

ax = fig.add_subplot(1,1,1, projection='3d')
ax.scatter3D(batsman_record_df['avg'], batsman_record_df['runs'], batsman_record_df['strike_rate'], c='red')
ax.set_title('Batsman Performance')
ax.set_xlabel('Average')
ax.set_ylabel('Runs')
ax.set_zlabel('Strike Rate')
fig.show()
````

### `np.meshgrid()`

It is used to create **coordinate matrices (grids)** from 1D arrays of x and y values.


#### Syntax
    X, Y = np.meshgrid(x, y)


| Parameter | Description                          |
|----------|--------------------------------------|
| x        | 1D array for x-axis                  |
| y        | 1D array for y-axis                  |



#### What it does

- Converts **1D arrays → 2D grid matrices**  
- Generates all possible combinations of x and y  
- Required for **3D surface plots**  


#### Output

- `X` → repeated rows of x values  
- `Y` → repeated columns of y values  


#### Shape Rule

```text
If x has length m and y has length n:

X.shape = (n, m)
Y.shape = (n, m)


### 3D Surface Plot (`ax.plot_surface()`)

It is used to visualize a **continuous surface (z values over x and y grid)** in 3D space.


#### Syntax
    ax.plot_surface(X, Y, Z)


| Parameter | Description                          |
|----------|--------------------------------------|
| X        | Grid of x values (2D array)          |
| Y        | Grid of y values (2D array)          |
| Z        | Corresponding z values (2D array)    |
| cmap     | Colormap for surface                 |
| edgecolor| Color of surface edges               |
| alpha    | Transparency                         |


#### Requirements

- X, Y, Z must be **2D arrays (meshgrid format)**  
- Shapes of X, Y, Z must be **same**  


#### Steps

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z)

### `plt.imshow()`

It is used to display **2D data as an image** (matrix → color representation).


#### Syntax
    plt.imshow(X)


| Parameter | Description                                      | Default | Possible Values                  |
|----------|--------------------------------------------------|---------|----------------------------------|
| X        | 2D array (image or matrix)                       | —       | array-like                       |
| cmap     | Colormap (color scheme)                          | None    | 'viridis', 'gray', 'hot', etc.   |
| interpolation | Method to display pixels                    | None    | 'nearest', 'bilinear', etc.      |
| aspect   | Aspect ratio                                     | 'equal' | 'auto', 'equal'                  |
| vmin     | Minimum value for color scaling                  | None    | float                            |
| vmax     | Maximum value for color scaling                  | None    | float                            |


#### What it does

- Converts **numerical matrix → colored image**  
- Each value → mapped to a color  
- Commonly used for:
  - heatmaps  
  - image data  
  - matrices  


#### Key Concepts

- Rows → vertical axis  
- Columns → horizontal axis  
- Values → colors  


#### Key Points

- Requires **2D data**  
- Works with **NumPy arrays / Pandas values**  
- Often used with `cmap` for better visualization  
- Can add colorbar using `plt.colorbar()`  


#### Key Insight

- `imshow()` = **visual representation of matrix values**  


#### Difference from Other Plots

| Plot Type   | Representation          |
|-------------|------------------------|
| Scatter     | Points                 |
| Line        | Continuous line        |
| Surface     | 3D shape               |
| Contour     | Lines/levels           |
| Imshow      | Color grid (image)     |

### `DataFrame.plot()` (Pandas Plot Function)

It is used to create **quick visualizations directly from a Pandas DataFrame or Series** using Matplotlib internally.


#### Syntax
    df.plot(kind='line')


| Parameter | Description                                      | Default | Possible Values                          |
|----------|--------------------------------------------------|---------|------------------------------------------|
| kind     | Type of plot                                     | 'line'  | 'line', 'bar', 'barh', 'hist', 'box', 'kde', 'area', 'scatter', 'pie' |
| x        | Column for x-axis                                | None    | column name                              |
| y        | Column(s) for y-axis                             | None    | column name, list                        |
| figsize  | Size of plot                                     | None    | tuple                                    |
| color    | Color(s)                                         | None    | color, list                              |
| title    | Title of plot                                    | None    | str                                      |
| grid     | Show grid                                        | False   | True, False                              |
| legend   | Show legend                                      | True    | True, False                              |


#### Common Plot Types

| Kind      | Description                          |
|-----------|--------------------------------------|
| line      | Line plot                            |
| bar       | Vertical bar chart                   |
| barh      | Horizontal bar chart                 |
| hist      | Histogram                            |
| box       | Box plot                             |
| kde       | Density plot                         |
| area      | Area plot                            |
| scatter   | Scatter plot                         |
| pie       | Pie chart                            |


#### Key Points

- Built on top of **Matplotlib**  
- Very **quick and concise** plotting  
- Automatically uses:
  - index → x-axis  
  - columns → y-axis  
- Supports multiple columns in one plot  


#### Key Insight

- `df.plot()` = **shortcut for Matplotlib plots**  
- Less control but faster  


#### When to Use

- Quick **EDA (Exploratory Data Analysis)**  
- Basic visualizations  
- When you don’t need heavy customization
