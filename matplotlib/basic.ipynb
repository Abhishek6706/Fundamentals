### `matplotlib.pyplot` Library

`matplotlib.pyplot` is used to create **visualizations and plots** in Python.

### 2D Line Plot (`plt.plot()`)

It is used to plot a **line graph between two variables (x and y)** in 2D space.


#### Syntax
    plt.plot(x, y, fmt)


| Parameter | Description                          | Default | Possible Values                  |
|----------|--------------------------------------|---------|----------------------------------|
| x        | X-axis data                          | —       | list, array, Series              |
| y        | Y-axis data                          | —       | list, array, Series              |
| fmt      | Format string (optional)             | None    | 'r-', 'bo--', etc.               |
| label    | Label for legend                     | None    | str                              |
| linewidth| Width of line                        | 1       | float                            |
| linestyle| Style of line                        | '-'     | '-', '--', ':', '-.'             |
| marker   | Marker style                         | None    | 'o', 's', '^', etc.              |
| color    | Line color                           | None    | 'r', 'g', 'b', etc.              |


#### Key Points

- Represents relationship between **two variables (x vs y)**  
- Default plot type in `matplotlib.pyplot`  
- Supports multiple lines in same plot  
- Can be customized with markers, colors, styles  
- Widely used for **trend analysis**

### Scatter Plot (`plt.scatter()`)

It is used to visualize the **relationship between two numerical variables** using points in 2D space.


#### Syntax
    plt.scatter(x, y, s=None, c=None, marker=None)


| Parameter | Description                                      | Default | Possible Values                  |
|----------|--------------------------------------------------|---------|----------------------------------|
| x        | Values for x-axis                                | —       | list, array, Series              |
| y        | Values for y-axis                                | —       | list, array, Series              |
| s        | Size of points                                   | None    | int, array                       |
| c        | Color of points                                  | None    | color name, list, array          |
| marker   | Shape of points                                  | 'o'     | 'o', 's', '^', 'D', etc.         |
| cmap     | Colormap (for multiple colors)                   | None    | 'viridis', 'plasma', etc.        |
| alpha    | Transparency of points                           | 1.0     | 0 to 1                           |
| label    | Label for legend                                 | None    | str                              |


#### Key Points

- Used to show **correlation/relationship between variables**  
- Each point represents a **data observation**  
- Useful for detecting:
  - patterns  
  - clusters  
  - outliers  
- Supports color, size, and transparency customization

### Bar Graph (`plt.bar()`)

It is used to represent **categorical data with rectangular bars**, where the height of each bar indicates the value.


#### Syntax
    plt.bar(x, height, width=0.8)


| Parameter | Description                                      | Default | Possible Values                  |
|----------|--------------------------------------------------|---------|----------------------------------|
| x        | Categories / positions on x-axis                 | —       | list, array, Series              |
| height   | Values corresponding to each category            | —       | list, array, Series              |
| width    | Width of each bar                               | 0.8     | float                            |
| color    | Color of bars                                   | None    | 'red', '#123abc', list           |
| label    | Label for legend                                | None    | str                              |
| align    | Alignment of bars                               | 'center'| 'center', 'edge'                 |
| edgecolor| Border color of bars                            | None    | color                            |
| linewidth| Thickness of bar border                         | None    | float                            |
| alpha    | Transparency of bars                            | 1.0     | 0 to 1                           |

#### Horizontal Bar Graph Syntax
    plt.barh(y, width)

```python
import numpy as np
```

```python
student = np.array([10,15,29,20,25])
choice = np.array(['Red', 'Orange', 'Grey', 'Black', 'Pink'])
```

```python
plt.bar(choice, student, color=choice)
plt.xlabel('Color')
plt.ylabel('Number of Students')
plt.title('Favourite color')
plt.show()
```

#### Horizontal Bar Graph

```python
plt.barh(choice, student, color=choice)
plt.ylabel('Color')
plt.xlabel('Number of Students')
plt.title('Favourite color')
plt.show()
```

### `plt.xticks()`

It is used to **control the tick positions and labels on the x-axis**.


#### Syntax
    plt.xticks(ticks=None, labels=None, rotation=0)


| Parameter | Description                                      | Default | Possible Values                  |
|----------|--------------------------------------------------|---------|----------------------------------|
| ticks    | Positions of ticks on x-axis                     | None    | list, array                      |
| labels   | Labels corresponding to ticks                    | None    | list of strings                  |
| rotation | Rotation angle of labels                         | 0       | int (degrees)                    |


#### What it does

- Sets **where ticks appear** on x-axis  
- Sets **what labels are shown**  
- Helps in customizing axis readability  


#### Common Use Case

- When x values are numeric (0,1,2,3...)  
- But you want to show **category names instead**


#### Key Points

- `ticks` → positions (numbers)  
- `labels` → names (categories)  
- Must have **same length**  
- Improves visualization clarity  


#### Rotation

- Used when labels overlap  
- Common values: `45`, `90`  


#### Key Insight

- Without `xticks()` → default numeric indexing  
- With `xticks()` → meaningful category labels  


### Plot the runs per season per batsman in a bar graph

```python
plt.bar(np.arange(batsman_season_record_df.shape[0])-0.25, batsman_season_record_df['2015'], width=0.25)
plt.bar(np.arange(batsman_season_record_df.shape[0]), batsman_season_record_df['2016'], width= 0.25)
plt.bar(np.arange(batsman_season_record_df.shape[0])+0.25, batsman_season_record_df['2017'], width=0.25)

plt.xticks(ticks=np.arange(batsman_season_record_df.shape[0]), labels=batsman_season_record_df['batsman'])

plt.show()
```

### Stacked Bar Chart (`plt.bar()` with `bottom`)

It is used to represent **multiple values stacked on top of each other** for each category.


#### Key Idea

- Use the `bottom` parameter to **stack bars vertically**
- Each new bar is placed **on top of the previous one**

```python
batsman_season_record_df
```

```python
plt.bar(batsman_season_record_df['batsman'], batsman_season_record_df['2015'], color='red', label='2015 runs')
plt.bar(batsman_season_record_df['batsman'], batsman_season_record_df['2016'], bottom= batsman_season_record_df['2015'], color='green', label='2016 runs')
plt.bar(batsman_season_record_df['batsman'], batsman_season_record_df['2017'], bottom=(batsman_season_record_df['2015'] +  batsman_season_record_df['2016']), color='blue', label='2017 runs')

plt.legend()

plt.xlabel('Batsman Name')
plt.ylabel('Total Runs')

plt.title('Overall Runs Distribution')

plt.show()
```

### Histogram (`plt.hist()`)

It is used to represent the **distribution of numerical data** by dividing it into intervals (bins).


#### Syntax
    plt.hist(x, bins=10)


| Parameter | Description                                      | Default | Possible Values                  |
|----------|--------------------------------------------------|---------|----------------------------------|
| x        | Input numerical data                             | —       | list, array, Series              |
| bins     | Number of intervals                              | 10      | int, list                        |
| color    | Color of bars                                   | None    | 'blue', '#123abc'                |
| edgecolor| Border color of bars                            | None    | color                            |
| alpha    | Transparency                                    | 1.0     | 0 to 1                           |
| density  | Normalize histogram                             | False   | True, False                      |
| histtype | Type of histogram                               | 'bar'   | 'bar', 'step', 'stepfilled'      |
| cumulative | Cumulative distribution                       | False   | True, False                      |


#### Key Concepts

- **Bins** → intervals that group data  
- Height of bar → **frequency (count)**  
- Shows **data distribution pattern**  


#### Key Points

- Used for **continuous numerical data**  
- Helps identify:
  - distribution shape  
  - skewness  
  - outliers  
- More bins → finer detail  
- Fewer bins → smoother overview  


#### Key Insight

- Histogram ≠ bar chart  
- Histogram → continuous data distribution  
- Bar chart → categorical comparison

### Pie Chart (`plt.pie()`)

It is used to represent **proportions or percentages of a whole** using circular slices.


#### Syntax
    plt.pie(x)


| Parameter   | Description                                      | Default | Possible Values                  |
|-------------|--------------------------------------------------|---------|----------------------------------|
| x           | Data values (sizes of slices)                    | —       | list, array, Series              |
| labels      | Labels for each slice                            | None    | list of strings                  |
| colors      | Colors of slices                                 | None    | list of colors                   |
| autopct     | Display percentage on slices                     | None    | '%1.1f%%', '%d%%'                |
| explode     | Offset slices outward                            | None    | list of floats (0–1)             |
| startangle  | Starting angle of chart                          | 0       | int                              |
| shadow      | Add shadow effect                                | False   | True, False                      |
| wedgeprops  | Customize slice properties                       | None    | dict                             |
| textprops   | Customize text properties                        | None    | dict                             |


#### Key Concepts

- Each slice represents a **portion of total**  
- Sum of all values = **100%**  
- Angle of slice ∝ value  


#### Important Parameters

- `autopct` → shows percentage  
- `explode` → highlights specific slice  
- `startangle` → rotates chart for better view  


#### Key Points

- Used for **proportion analysis**  
- Best for **few categories (≤5)**  
- Hard to interpret with many slices  
- Not ideal for precise comparison  


#### Key Insight

- Pie chart shows **relative contribution**, not exact values  

```python
gayle_175_match_details_df = pd.read_csv('gayle-175.csv')
```

```python
gayle_175_match_details_df.head()
```

```python
gayle_175_match_details_df.shape
```

### `plt.savefig()`

It is used to **save the current figure (plot) as an image/file**.


#### Syntax
    plt.savefig(fname, dpi=None)


| Parameter   | Description                                      | Default | Possible Values                  |
|-------------|--------------------------------------------------|---------|----------------------------------|
| fname       | File name or path to save plot                   | —       | 'plot.png', 'chart.pdf', etc.    |
| dpi         | Resolution (dots per inch)                       | None    | int (e.g., 100, 300)             |
| format      | File format                                      | Auto    | 'png', 'jpg', 'pdf', 'svg'       |
| bbox_inches | Adjust bounding box                              | None    | 'tight'                          |
| transparent | Transparent background                           | False   | True, False                      |
| facecolor   | Background color                                 | None    | color                            |
| edgecolor   | Edge color                                       | None    | color                            |


#### Key Points

- Saves plot to **file instead of displaying only**  
- Supports multiple formats (PNG, JPG, PDF, SVG)  
- Use **high dpi (e.g., 300)** for better quality  
- `bbox_inches='tight'` removes extra whitespace  


#### Important Notes

- Call `savefig()` **before `plt.show()`**  
- Otherwise, saved image may be blank  


#### Key Insight

- `show()` → display plot  
- `savefig()` → store plot  


```python
plt.pie(gayle_175_match_details_df['batsman_runs'], labels=gayle_175_match_details_df['batsman'], autopct='%0.1f%%', explode=[0.1,0,0,0,0,0])

plt.savefig('gayle_175_data_distribution.png')
```
