### What is numpy?

NumPy is the fundamental package for scientific computing in Python. 

It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

At the core of the NumPy package, is the ndarray object. This encapsulates n-dimensional arrays of homogeneous data types

### Numpy Arrays Vs Python Sequences

- NumPy arrays have a fixed size at creation, unlike Python lists (which can grow dynamically). Changing the size of an ndarray will create a new array and delete the original.

- The elements in a NumPy array are all required to be of the same data type, and thus will be the same size in memory.

- NumPy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically, such operations are executed more efficiently and with less code than is possible using Python’s built-in sequences.

### Data Structures: Scalar, Vector, Matrix, and Tensor

In NumPy, all data structures are instances of the ndarray class, categorized by their rank (number of dimensions):

- **Scalar (0D)**: A single numerical value with no dimensions (Rank 0).
   - Example: A single temperature reading like 25.
    - Shape: ()

- **Vector (1D)**: A simple array of numbers arranged in a single dimension (Rank 1).

    - Example: A list of prices [100, 200, 300].
    - Shape: (n,)

- **Matrix (2D)**: A collection of numbers arranged in a grid with rows and columns (Rank 2).

    - Example: A spreadsheet or a grayscale image.
    - Shape: (rows, columns)

- **Tensor (3D+)**: The generalized term for arrays with 3 or more dimensions (Rank 3+).
    - Example: A color image (Height, Width, Channels) or a video.
    - Shape: (d1, d2, d3, ...)

### np.arange() Function
The arange() function returns an ndarray with evenly spaced values within a given interval. 

It is the NumPy equivalent of Python's built-in range() function but returns an array instead of a list.

- **Syntax**: np.arange([start], stop, [step], dtype=None)

- **Key Characteristics**:

    - **Start**: The interval includes this value (default is 0).
    - **Stop**: The interval does not include this value (exclusive).
    - **Step**: The spacing between values (default is 1). This can be a floating-point number, unlike Python's range().

```python
f= np.arange(start=2, stop=10, step=3)
print(f)
```

```python
g= np.arange(1,5,2)
g
```

### np.reshape() Function
The reshape() function changes the shape of an array without changing its data. 

- **Syntax**: array.reshape(shape) 
- **Key Rules**:
    - **Total Elements**: The new shape must be compatible with the original shape. The total number of elements (product of dimensions) must remain the same.
        - *Example*: An array with $12$ elements can be reshaped to $(3, 4)$, $(6, 2)$, or $(2, 2, 3)$, but not $(5, 2)$.
    - **The -1 Shortcut**: You can pass -1 as one of the dimensions. NumPy will automatically calculate the correct value for that dimension based on the remaining elements.
    - **Memory**: Reshaping usually returns a view of the original array, meaning it does not copy the data in memory, making it very efficient.

```python
h = np.arange(1,13).reshape(3,-1)
h
```

```python
j = np.arange(1,19,3).reshape(2,3)
j
```

### np.linspace() Function
The linspace() (linear space) function returns an array of evenly spaced numbers over a specified interval. 

While it is similar to arange(), it is specifically designed for cases where you know the number of elements you want rather than the step size.

- **Syntax**: np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)

- **Key Characteristics**:

    - **Start and Stop**: Defines the range of the sequence. Unlike arange(), the stop value is included by default (endpoint=True).

    - **Num**: Specifies the total number of samples to generate. If not provided, it defaults to 50.

    - **Precision**: It calculates the step size automatically to ensure exactly the requested number of elements, making it ideal for creating coordinate grids and plotting functions.

    - **Step Calculation**: If retstep is set to True, the function returns both the array and the calculated step size used between the values.

```python
a = np.linspace(start=10, stop = 100, num = 30, retstep=True)
a
```

```python
a = np.linspace(start=10, stop = 100, num = 30, retstep=False)
a
```

### np.identity() Function
The identity() function creates a square identity matrix of a given size. An identity matrix is a special square matrix with ones on the main diagonal and zeros elsewhere.

- **Syntax**: np.identity(n, dtype=None)
- **Key Characteristics**:
    -  **Square Matrix**: Unlike zeros or ones, identity only takes a single integer n as an argument because the result is always an $n \times n$ square matrix.
    - **Main Diagonal**: The "main diagonal" (from top-left to bottom-right) is populated with 1.0, while all other elements are 0.0.
    - **Mathematical Role**: In linear algebra, multiplying any matrix $A$ by the identity matrix $I$ results in the original matrix $A$ (similar to multiplying a number by $1$).
    - **Default Dtype**: Defaults to float64. You can change this to int or other types using the dtype parameter.

```python
a = np.identity(n = 4, dtype = int)
a
```

```python
a = np.identity(n = 4)
a
```

### NumPy Array Attributes
Attributes in NumPy allow you to inspect the intrinsic properties of an ndarray without calling a function. These are essential for debugging and understanding the structure of your data.
- **ndarray.ndim**: Returns the number of dimensions (axes) of the array.   
    - Scalar: 0
    - Vector: 1
    - Matrix: 2
- ndarray.shape: Returns a tuple of integers indicating the size of the array in each dimension.
    - Example: For a matrix with $n$ rows and $m$ columns, the shape will be (n, m).
- ndarray.size: Returns the total number of elements in the array.          
    - Calculation: This is equal to the product of the elements of its shape.
- ndarray.dtype: Returns an object describing the type of the elements in the array.

    - Example: An array of type float64 (8 bytes) has an itemsize of 8.
    
- ndarray.nbytes: Returns the total bytes consumed by the elements of the array.
    - Calculation: size $\times$ itemsize.

### astype() Function
The astype() function is the primary method used to cast or convert a NumPy array from one data type to another. Since NumPy arrays are homogeneous, this function creates a new copy of the array with the specified dtype.

- **Syntax**: ndarray.astype(dtype, casting='unsafe', copy=True)

- **Key Characteristics**:

    - **Type Conversion**: It allows you to convert between fundamentally different types (e.g., converting Floats to Integers, or Strings to Numbers).

     - **Data Truncation**: When converting from float to int, NumPy will truncate the decimal part (e.g., 3.9 becomes 3), not round it.

    - **Memory Management**: By default, astype() always returns a new copy of the data, even if the destination type is the same as the source type.

    - **Error Handling**: If a conversion is impossible (e.g., converting the string "abc" to an int), it will raise a ValueError.

```python
a = np.array([[[2,3],[4,5]],[[1,2],[3,4]]])
a.dtype
```

```python
a = np.array([[[2,3],[4,5]],[[1,2],[3,4]]]).astype(dtype=np.float32)
print(a)
a.dtype
```

```python
a = np.array([[[2,3],[4,5]],[[1,2],[3,4]]]).astype(dtype=np.int16)
print(a)
a.dtype
```

### np.var() Function
The var() function calculates the Variance of the elements in an array. Variance is a statistical measurement of the spread between numbers in a data set; it measures how far each number in the set is from the mean.

- **Syntax**: np.var(a, axis=None, dtype=None, ddof=0, keepdims=False)
- **Key Characteristics**:
    - **Calculation**: It is the average of the squared deviations from the mean.
    - **Mathematical Formula**: The variance ($\sigma^2$) is calculated as:$$\sigma^2 = \frac{\sum (x_i - \mu)^2}{N}$$where $x_i$ is each element, $\mu$ is the mean, and $N$ is the number of elements.
    - **Relationship to Standard Deviation**: The variance is simply the square of the standard deviation ($\sigma^2 = \text{std}^2$).
    - **Axis Parameter**: Like other aggregation functions, you can calculate variance for the entire array, column-wise (axis=0), or row-wise (axis=1).

```python
np.std(a1, axis=0)
```

```python
np.var(a1, axis = 0)
```

```python
np.std(a2)
```

```python
np.var(a2)
```

## np.round() Function
The round() function rounds an array to the nearest integer or a specified number of decimals.

- **Syntax**: np.round(a, decimals=0)

- **Key Characteristics**:

    - **Precision**: The decimals parameter allows you to round to a specific decimal place (e.g., decimals=2).

    - **Rounding to Even**: NumPy uses "Banker's Rounding." If a number is exactly halfway between two integers (like 0.5 or 1.5), it rounds to the nearest even integer.

    - *Example*: np.round(0.5) is 0.0, and np.round(1.5) is 2.0.

## np.floor() Function

The floor() function returns the largest integer less than or equal to each element (it always rounds down toward negative infinity).

- **Syntax**: np.floor(a)
- **Key Characteristics**:
    - **Direction**: Unlike rounding, it always moves to the left on the number line.
    - **Negative Numbers**: For -1.1, the floor is -2.0 (because $-2$ is smaller than $-1.1$).

## np.ceil() Function

The ceil() (short for "ceiling") function returns the smallest integer greater than or equal to each element (it always rounds up toward positive infinity).
- **Syntax**: np.ceil(a)
- **Key Characteristics**:
     - **Direction**: It always moves to the right on the number line.
     - **Negative Numbers**: For -1.9, the ceil is -1.0 (because $-1$ is larger than $-1.9$).

## Indexing and Slicing

Indexing and slicing allow you to access specific elements or sub-sections of an array. In NumPy, this is done using the bracket [] notation.

## Vector Indexing and Slicing (1D)
For a 1D array, indexing works exactly like a standard Python list.

- **Indexing**: Access a single element using its position (starting from 0).

    - *Example*: arr[0] is the first element, arr[-1] is the last.

- **Slicing**: Access a range of elements using the [start:stop:step] syntax.

    - *Note*: The stop index is exclusive.

## Matrix Indexing and Slicing (2D)
In a 2D array, you use two indices separated by a comma: [row, column].

- **Individual Element**: arr[1, 2] accesses the element at row index 1 and column index 2.

- **Slicing Rows/Columns**:

    - *Full Row*: arr[1, :] (Gets all columns in the second row).

    - *Full Column*: arr[:, 2] (Gets all rows in the third column).

    - *Sub-matrix*: arr[0:2, 1:3] (Gets rows 0-1 and columns 1-2).

## Tensor Indexing and Slicing (3D+)
For Tensors, you add more indices separated by commas: [depth, row, column].

- **Accessing a Plane**: arr[0, :, :] (Selects the first 2D matrix/slice in a 3D stack).

- **Accessing a Vector within a Tensor**: arr[0, 1, :] (Selects the second row of the first matrix).

- **The Ellipsis (...)**: If you have many dimensions and want to select all remaining axes, you can use ....

    - *Example*: arr[0, ...] is equivalent to arr[0, :, :].

## Transpose (.T Attribute)

The Transpose of a matrix is an operator that flips a matrix over its main diagonal, switching its row and column indices.
- **Syntax**: array.T or np.transpose(array)
- **Key Characteristics**:
    - **Dimension Swap**: For a 2D array of shape $(n, m)$, the transpose will have a shape of $(m, n)$.
    - **1D Arrays**: Transposing a 1D array (vector) has no effect because it only has one axis. To turn a 1D vector into a column vector, you must first reshape it to 2D.
    - **View vs. Copy**: Like reshape, .T usually returns a view of the original array, making it an $O(1)$ operation in terms of time and memory.
    - **Higher Dimensions**: For tensors, .T reverses the order of the axes (e.g., a shape of $(2, 3, 4)$ becomes $(4, 3, 2)$).

## np.ravel() Function
The ravel() function is used to flatten a multi-dimensional array into a 1-dimensional array.

- **Syntax**: np.ravel(array, order='C')

- **Key Characteristics**:

    - *Flattening*: It unrolls the array into a single line of data.

    - *Memory Efficiency*: Unlike ndarray.flatten(), which always returns a copy, ravel() returns a view whenever possible. Modifying the raveled array may modify the original array.

    - *Order Parameter*:

        - **'C'**: (Default) Flattens row-by-row (C-style).

        - **'F'**: Flattens column-by-column (Fortran-style).

# Stacking in NumPy

Stacking is the process of joining a sequence of arrays along a new or existing axis. It is essential for combining datasets, such as merging multiple image channels or adding new observations to a data matrix.


## np.vstack() (Vertical Stack)

The vstack() function stacks arrays in sequence vertically.

- **Syntax**: np.vstack((array1, array2, ...))
- **Key Characteristics**:
    - *Requirement*: The arrays must have the same number of columns.
    - *Dimensionality*: Increases the number of rows.
    - *Use Case*: Adding new rows (samples) to an existing dataset.

```python
a1= np.arange(12).reshape(3,4)
a2 = np.arange(12,24).reshape(3,4)
```

```python
a1
```

```python
a2
```

```python
np.vstack(tup=(a1,a2))
```

```python
np.vstack(tup=(a2,a1))
```

## np.hstack() (Horizontal Stack)
The hstack() function stacks arrays in sequence horizontally (column-wise).

- **Syntax**: np.hstack((array1, array2, ...))

- **Key Characteristics**:

    - *Requirement*: The arrays must have the same number of rows.

    - *Dimensionality*: Increases the number of columns.

    - *Use Case*: Adding new features (columns) to an existing dataset.

```python
np.hstack(tup=(a1,a2))
```

```python
a3 = np.hstack(tup=(a2,a1))
a3
```

# Splitting in NumPy
Splitting is the analytical opposite of stacking. It allows you to break a single array into multiple smaller sub-arrays along a specified axis.

## np.vsplit() (Vertical Split)
The vsplit() function splits an array into multiple sub-arrays row-wise. It only works on arrays with at least 2 dimensions.

- **Syntax**: np.vsplit(array, indices_or_sections)

- **Key Characteristics**:

    - *Sections*: If an integer is passed, the array is split into that many equal parts.

    - *Indices*: If a list of integers is passed, it splits at those specific row indices.

    - *Requirement*: The number of rows must be divisible by the number of sections.

```python
a3
```

```python
np.vsplit(a3,3)
```

```python
np.vsplit(a3,[0,1])
```

## np.hsplit() (Horizontal Split)
The hsplit() function splits an array into multiple sub-arrays column-wise.

- **Syntax**: np.hsplit(array, indices_or_sections)

- **Key Characteristics**:

    - *Direction*: It cuts the array "down the middle" of the columns.

    - *Use Case*: Separating a target label column from a feature matrix in machine learning.

    - *Requirement*: The number of columns must be divisible by the number of sections.

```python
a3
```

```python
np.hsplit(a3, 2)
```

```python
np.hsplit(a3,[2,4,5])
```
